```python
Busket = df[df['Date'].duplicated(keep=False)]
df['Group'] = df.groupby('Date')['Company'].transform(lambda x: ','.join(x))

df2 = df[['Date','Group']].drop_duplicates()
count = Counter()
for row in df['Group']:
  row_list = row.split(',')
  count.update(Counter(combinations(row_list,2)))
for key, value in count.most_common(10):
  print(key,value)
```
